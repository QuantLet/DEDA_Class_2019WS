```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
```

## 1. Price data

### 1. - 1) Raw price data

```python
price = pd.read_csv('1718hourlyPrice_withZEC.csv', index_col = 'time')
price.index = pd.to_datetime(price.index)


# fig = go.Figure()
# for coin in price.columns:
#     fig.add_trace(go.Scatter(x=price.index, y=price[coin],name = coin))

# fig.update_layout(title="Hourly Price")
```

### 1. - 2) Log price data

```python
LogPrice = np.log(price)
LogPrice.index = pd.to_datetime(LogPrice.index)


# fig = go.Figure()
# for coin in LogPrice.columns:
#     fig.add_trace(go.Scatter(x=LogPrice.index, y=LogPrice[coin],name = coin))

# fig.update_layout(title="Log Hourly Price")
```

### 1. - 3) Standardized price data

```python
StdPrice = (price - price.mean(0)) / price.std(0)
StdPrice.index = pd.to_datetime(StdPrice.index)


# fig = go.Figure()
# for coin in StdPrice.columns:
#     fig.add_trace(go.Scatter(x=StdPrice.index, y=StdPrice[coin],name = coin))

# fig.update_layout(title="Standardized Hourly Price")
```

#### Export the graph as a png file

```python
%matplotlib inline 
from matplotlib import pyplot as plt
import matplotlib
from matplotlib.dates import MonthLocator, DateFormatter
import matplotlib.dates as mdates


fig =  plt.figure(figsize = (12, 18))
ax1 = fig.add_subplot(311)

for coin in price.columns:
    ax1.plot(price.index, price[coin], label = coin, linewidth=1)

ax1.grid(axis = 'y')



date = price.index.to_pydatetime()

ax1.set_xlim([date[0], date[-1]])

ax1.xaxis.set_major_locator(MonthLocator())
ax1.xaxis.set_major_formatter(DateFormatter("%b %Y"))
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.spines["bottom"].set_visible(False)
ax1.spines["left"].set_visible(False)
ax1.legend(loc= 'upper right', prop={'size': 8}, frameon=False)
ax1.title.set_text("Hourly Price")


ax2 = fig.add_subplot(312)

for coin in LogPrice.columns:
    ax2.plot(LogPrice.index, LogPrice[coin], label = coin, linewidth=1)

ax2.grid(axis = 'y')

date = LogPrice.index.to_pydatetime()

ax2.set_xlim([date[0], date[-1]])

ax2.xaxis.set_major_locator(MonthLocator())
ax2.xaxis.set_major_formatter(DateFormatter("%b %Y"))
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.spines["bottom"].set_visible(False)
ax2.spines["left"].set_visible(False)

ax2.legend(loc= 'upper right', prop={'size': 8}, frameon=False)
ax2.title.set_text("Log Hourly Price")


ax3 = fig.add_subplot(313)

for coin in StdPrice.columns:
    ax3.plot(StdPrice.index, StdPrice[coin], label = coin, linewidth=1)

ax3.grid(axis = 'y')
ax3.legend(prop={'size': 8}, frameon=False)

ax3.title.set_text("Standardized Hourly Price")


date = StdPrice.index.to_pydatetime()

ax3.set_xlim([date[0], date[-1]])

ax3.xaxis.set_major_locator(MonthLocator())
ax3.xaxis.set_major_formatter(DateFormatter("%b %Y"))
ax3.spines["top"].set_visible(False)
ax3.spines["right"].set_visible(False)
ax3.spines["bottom"].set_visible(False)
ax3.spines["left"].set_visible(False)


plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()


# fig.savefig('Price data analysis.png')
```
